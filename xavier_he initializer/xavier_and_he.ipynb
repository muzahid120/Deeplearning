```python
import pandas as pd 
import numpy as np 
from sklearn.datasets import make_moons
import tensorflow 
from tensorflow import keras 
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import seaborn as sns
from keras.initializers import glorot_uniform,he_uniform
```

```python
gn=glorot_uniform()
```

```python
x,y=make_moons(n_samples=100,noise=0.4,random_state=2)
```

```python
sns.scatterplot(x=x[:,0],y=x[:,1],hue=y)
```

```python
model1=Sequential()
model1.add(Dense(20,activation='tanh',input_dim=2,kernel_initializer=gn))
model1.add(Dense(20,activation='tanh',kernel_initializer=gn))
model1.add(Dense(20,activation='tanh',kernel_initializer=gn))
model1.add(Dense(20,activation='tanh',kernel_initializer=gn))
model1.add(Dense(1,activation='sigmoid',kernel_initializer=gn))
```

```python
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history=model1.fit(x,y,epochs=200,validation_split=0.2,verbose=1)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x,y.astype('int'),clf=model1,legend=2)
```

```python
plt.plot(history.history['val_loss'])
plt.plot(history.history['loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

```python
he=he_uniform()
```
