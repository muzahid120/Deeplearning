```python
import pandas as pd 
import numpy as np 
from sklearn.datasets import make_moons
import tensorflow 
from tensorflow import keras 
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
x,y=make_moons(n_samples=100,noise=0.3,random_state=2)
```

```python
model1=Sequential()
model1.add(Dense(20,activation='tanh',input_dim=2))
model1.add(Dense(20,activation='tanh'))
model1.add(Dense(20,activation='tanh'))
model1.add(Dense(20,activation='tanh'))
model1.add(Dense(1,activation='sigmoid'))
```

```python
initial_weight=model1.get_weights()
initial_weight
```

```python
initial_weight[0]=np.random.randn(model1.get_weights()[0].shape[0],model1.get_weights()[0].shape[1])*2
initial_weight[1]=np.zeros(model1.get_weights()[1].shape)
initial_weight[2]=np.random.randn(model1.get_weights()[2].shape[0],model1.get_weights()[2].shape[1])*2
initial_weight[3]=np.zeros(model1.get_weights()[3].shape)
initial_weight[4]=np.random.randn(model1.get_weights()[4].shape[0],model1.get_weights()[4].shape[1])*2
initial_weight[5]=np.zeros(model1.get_weights()[5].shape)
initial_weight[6]=np.random.randn(model1.get_weights()[6].shape[0],model1.get_weights()[6].shape[1])*2
initial_weight[7]=np.zeros(model1.get_weights()[7].shape)
initial_weight[8]=np.random.randn(model1.get_weights()[8].shape[0],model1.get_weights()[8].shape[1])*2
initial_weight[9]=np.zeros(model1.get_weights()[9].shape)
```

```python
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history=model1.fit(x,y,epochs=200,validation_split=0.2,verbose=1)
```
