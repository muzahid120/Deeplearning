```python
df.drop(columns=['RowNumber', 'CustomerId', 'Surname'],inplace=True)
```

```python
df=pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
df
```

```python
def step(x):
    return 1 if x==True else 0
```

```python
df['Geography_Germany']=df['Geography_Germany'].apply(step)
df['Geography_Spain']=df['Geography_Spain'].apply(step)
df['Gender_Male']=df['Gender_Male'].apply(step)
```

```python
df
```

```python
x=df.drop(columns=['Exited'])

y=df['Exited']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
import tensorflow
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# simple neural_network model

model=Sequential()
model.add(Dense(3,activation='sigmoid',input_dim=11))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
histoty=model.fit(x_train_scaled,y_train,epochs=200)
```

```python
y_pred=model.predict(x_test_scaled)
```

```python
import numpy as np 
y_pred=np.where(y_pred>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
model2=Sequential()
model2.add(Dense(11,activation='relu',input_dim=11))
model2.add(Dense(11,activation='relu'))
model2.add(Dense(1,activation='sigmoid'))
```

```python
model2.summary()
```

```python
model2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history1=model2.fit(x_train_scaled,y_train,epochs=200,validation_split=0.2,validation_data=(x_test_scaled,y_test),verbose=1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```

```python
plt.plot(history1.history['accuracy'])
plt.plot(history1.history['val_accuracy'])
```
