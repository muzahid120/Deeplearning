```python
import pandas as pd 
import numpy as np 
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
x,y=make_moons(n_samples=200,noise=0.2,random_state=2)
len(x[:,1])
```

```python
def emwa(x,alpha=0.9):
    emwa_valus=np.zeros(x.shape[0])
    emwa_valus[0]=x[0]
    for i in range(1,len(x)):
        emwa_valus[i]=(alpha*x[i])+(1-alpha)*emwa_valus[i-1]
    return emwa_valus
```

```python
df=pd.DataFrame({'x1':x[:,0],'x2':x[:,1],'y':y})
```

```python
plt.scatter(df['x1'],df['x2'])
```

```python
df['emwa']=emwa(df['x1'],alpha=0.9)
```

```python

plt.plot(df['x1'],df['emwa'])
plt.show()
```

```python
df['ewm']=df['x1'].ewm(alpha=0.9).mean()
```
