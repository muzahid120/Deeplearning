```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import tensorflow 
from tensorflow import keras 
from keras.models import Sequential
from keras.layers import Dense,Dropout,BatchNormalization
from sklearn.datasets import make_circles
from keras.initializers import he_uniform
```

```python
x,y=make_circles(n_samples=200,noise=0.2,random_state=2)
```

```python
he=he_uniform()
```

```python
sns.scatterplot(x=x[:,0],y=x[:,1],hue=y)
```

```python
model=Sequential()
```

```python
model.add(Dense(3,activation='relu',input_dim=2))
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history=model.fit(x,y,epochs=100,validation_split=0.2,verbose=1)
```

```python
model1=Sequential()
```

```python
model1.add(Dense(3,activation='relu',input_dim=2))
model1.add(BatchNormalization())
model1.add(Dense(3,activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(3,activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(3,activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(3,activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(2,activation='relu'))
model1.add(Dense(1,activation='sigmoid'))
```

```python
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history=model1.fit(x,y,epochs=200,validation_split=0.2,verbose=1)
```

```python
plt.plot(history.history['val_accuracy'],color='green')
plt.plot(history.history['accuracy'],color='blue')
```
