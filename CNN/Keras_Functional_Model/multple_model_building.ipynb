```python
import tensorflow
from tensorflow import keras

from keras.models import Model
from keras.layers import *
```

```python
from re import X
# input_layers
input_A=Input(shape=(32,))
input_b=Input(shape=(128,))

#first_brance
x=Dense(64,activation='relu')(input_A)
x_1=Dense(32,activation='relu')(x)
x_2=Dense(16,activation='relu')(x_1)

#2nd_branch
y=Dense(64,activation='relu')(input_b)
y_1=Dense(32,activation='relu')(y)
y_2=Dense(16,activation='relu')(y_1)


#combine_the two branch
combine=concatenate([x_2,y_2])

# apply a FC layer and then a regression prediction on the
# combined outputs
z = Dense(2, activation="relu")(combine)
z1 = Dense(1, activation="linear")(z)


```

```python
model=Model(inputs=[input_A,input_b],outputs=z1)
```

```python
from keras.utils import plot_model
plot_model(model,show_layer_activations=True,show_trainable=True,show_layer_names=True,show_shapes=True,show_dtype=True)
```
